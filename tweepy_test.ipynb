```python
number_of_tweets = 100
tweets = []
likes = []
time = []
tweet_RT = []

cursor = tw.Cursor(api.user_timeline, id='elonmusk', tweet_mode="extended")  #extended for tweets > 140 Characters
for tweet in cursor.items(number_of_tweets):
    if (not tweet.retweeted) and ('@' not in tweet.full_text):
        tweets.append(tweet.full_text)  
        likes.append(tweet.favorite_count)
        time.append(tweet.created_at)
        tweet_RT.append(tweet.retweet_count)

df = pd.DataFrame({'Tweets':tweets,'Likes':likes,'Retweet':tweet_RT,'Time':time})
```

```python
def remove_url(txt):
    """Replace URLs found in a text string with nothing 
    (i.e. it will remove the URL from the string).

    Parameters
    ----------
    txt : string
        A text string that you want to parse and remove urls.

    Returns
    -------
    The same txt string with url's removed.
    """

    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt).split())
```

```python
def remove_url_th(txt):
    """Replace URLs found in a text string with nothing 
    (i.e. it will remove the URL from the string).

    Parameters
    ----------
    txt : string
        A text string that you want to parse and remove urls.

    Returns
    -------
    The same txt string with url's removed.
    """

    return " ".join(re.sub("([^\u0E00-\u0E7Fa-zA-Z' ]|^'|'$|''|(\w+:\/\/\S+))", "", txt).split())
```

```python
all_tweets_no_urls = [remove_url_th(tweet) for tweet in all_tweets]
all_tweets_no_urls[:20]
```
